```python
plt.figure(figsize= (20,10))
color=["#011638", "#7e2987", "#ef2026"]
labels= ["Diesel","Petrol", "Gas"]
sns.set_style("white")
plt.stackplot(data_auto["Year"], data_auto["Diesel"], data_auto["Petrol"], data_auto["Gas"],
              colors= color, edgecolor= 'None')

plt.xlabel("Years", fontsize= '15', weight="bold")
plt.ylabel("Number of Cars", fontsize= '15', weight="bold")

plt.xticks(data_auto["Year"], fontsize= '15', rotation= 90)
plt.yticks(fontsize= '15')

plt.legend(labels, loc= 'upper left', fontsize='15')
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize= '25', weight= "bold")
sns.despine() # removes the outer borders(Top and Right side) of the chart
plt.show()
```

- As per above Staked Bar Chart, there were very few cars in between the year 1982 and year 1990. So, its hard to compare
different fuels based engines.
- From early 90's there is gradual rise in the number of cars having engines based on Diesel and petrol.
- There is increase in Diesel based Engines from 1997, and from 2000 Diesel has dominated the Fuel market.
- Gas got hiked from year 2006 until 2008, and then sharp decline due to Global Stock Market crashed. Also the pattern is 
same with other fuels.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
data= pd.read_csv('C:/Users/Rohit/Desktop/Data Visualisation using Matplotlib Library/Python Users Survey data.csv')
```

```python
data
```

```python
data.info()
```

```python
import datetime
```

```python
data.Year = data.Year.astype(str)
```

```python
data.info()
```

```python
#Projecting Bar graph
fig, ax1= plt.subplots(figsize=(15,7))
ax1.bar(data['Year'], data['Participants'], color= 'g')
ax1.set_ylabel("Numbers of Participants",color='g', weight= "bold")
ax1.tick_params(axis='y', color='g', width=2, labelsize="large")

ax2= ax1.twinx() # Used to create 2 Y-axis having 1 common x-axis

# Project Line Graph using same x-axis and different y-axis
ax2.plot(data['Year'], data['Python Users'], color= 'r', marker= 'o')
ax2.set_ylim(0,1)
ax2.yaxis.set_major_formatter(PercentFormatter(xmax= 1.0))
ax2.set_ylabel("Python Users(%)", color='r', weight="bold")
ax2.tick_params(axis='y',color='r', width=2, labelsize="large")

plt.title(" KD Nuggets Survey of Python Users (2012 - 2019)", fontsize= '15', weight="bold")
plt.show()
```

- We can see from above Combo Chart, that Python Users have increased more that 40% since 2013 untill 2018. 
- Number of Participants are nearly constant between 2014 and 2018. And it is Second lowest in year 2019 with less than 2000 Participants.

# 2) Combining different Styles of Plots

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize= (15,10))
dataset = randn(100)

# Use distplot for combining plots, by default a kde over a histogram is shown
sns.distplot(dataset,bins=25)

plt.title('Combining different styles of Plotting', fontsize='17', weight='bold')
```

- The above graph shows the combination of Histogram and Kernal Density Estimation (KDE) of the dame Dataset 

```python
plt.figure(figsize= (15,10))

sns.distplot(dataset,bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},
             hist_kws={'color':'g','label':"HISTOGRAM"})
plt.legend()
plt.title('Combining different styles of Plotting', fontsize='17', weight='bold')
```
